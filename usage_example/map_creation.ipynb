```python
from shapely.geometry import Point
from worldmap import WorldMap
```

```python
map = WorldMap(
    boundary=[
        (0, 0),
        (0, 200),
        (300, 200),
        (300, 0),
    ],
    obstacles=[
        [
            (28, 20),
            (32, 20),
            (32, 199),
            (30, 200),
            (28, 199),
            (28, 20),
        ],
        [
            (300 - 28, 200 - 20),
            (300 - 32, 200 - 20),
            (300 - 32, 200 - 199),
            (300 - 30, 200 - 200),
            (300 - 28, 200 - 199),
            (300 - 28, 200 - 20),
        ],
        list(Point(70, 75).buffer(20).exterior.coords),
        list(Point(300 - 70, 200 - 75).buffer(20).exterior.coords),
        [
            (140, 10),
            (160, 10),
            (160, 80),
            (140, 80),
            (140, 10),
        ],
        [
            (300 - 160, 200 - 10),
            (300 - 140, 200 - 10),
            (300 - 140, 200 - 80),
            (300 - 160, 200 - 80),
            (300 - 160, 200 - 10),
        ],
        list(Point(14, 160).buffer(10).exterior.coords),
        [
            (60, 150),
            (70, 150),
            (70, 170),
            (90, 170),
            (90, 180),
            (60, 180),
            (60, 150),
        ],
        [
            (300 - 70, 200 - 150),
            (300 - 60, 200 - 150),
            (300 - 60, 200 - 180),
            (300 - 90, 200 - 180),
            (300 - 90, 200 - 170),
            (300 - 70, 200 - 170),
            (300 - 70, 200 - 150),
        ],
    ],
)
```

```python
map.visualize()
```

```python
map.save2json(file_name="map_name.json")
```
